```python
# dataset de mortes COVID19 
mortes = pd.read_csv('https://raw.githubusercontent.com/CSSEGISandData/COVID-19/master/csse_covid_19_data/csse_covid_19_time_series/time_series_covid19_deaths_global.csv')

# dataset de recuperação COVID19
recuperados = pd.read_csv('https://raw.githubusercontent.com/CSSEGISandData/COVID-19/master/csse_covid_19_data/csse_covid_19_time_series/time_series_covid19_recovered_global.csv')

display(mortes.head())
display(recuperados.head())
```

```python
recuperados.iloc[:,-272:-271].sum()
```

```python
display(casos_confirmados.groupby('Country/Region').sum().iloc[:,-1].sort_values(ascending = False).head())
display(mortes.groupby('Country/Region').sum().iloc[:,-1].sort_values(ascending = False).head())
display(recuperados.groupby('Country/Region').sum().iloc[:,-1].sort_values(ascending = False).head())
```

```python
print(casos_confirmados.shape)
print(mortes.shape)
print(recuperados.shape)
```

```python
def Ultima_data(df):
    return df.groupby('Country/Region').sum().iloc[:,-1]
```

```python
combinacao = [Ultima_data(casos_confirmados), Ultima_data(mortes), Ultima_data(recuperados)]
combinacao
```

```python
df_combinacao = pd.concat(combinacao, axis=1)
df_combinacao.columns = ['confirmados', 'mortes', 'recuperados']
df_combinacao
```

```python
resumo = df_combinacao.sum()
print((resumo['mortes'] / resumo['confirmados']) * 100)
print((resumo['mortes'] / (resumo['recuperados'] + resumo['mortes'])) * 100)
```

```python
resumo
```

```python
df_combinacao['letalidade'] = (df_combinacao.mortes / df_combinacao.confirmados) * 100
df_combinacao
```

```python
df_combinacao.info()
```

```python
soma_casos = [casos_confirmados.groupby('Country/Region').sum().iloc[:,5:].max(axis = 1), 
              mortes.groupby('Country/Region').sum().iloc[:,5:].max(axis = 1),
              recuperados.groupby('Country/Region').sum().iloc[:,5:].max(axis = 1)]
df_soma_casos = pd.concat(soma_casos, axis = 1)
df_soma_casos.columns = ['confirmados', 'mortes', 'recuperados']
df_soma_casos
```

```python
df_soma_casos['letalidade_1'] = df_soma_casos['mortes'] / df_soma_casos['confirmados'] * 100
df_soma_casos.head()
```

```python
df_soma_casos['letalidade_2'] = df_soma_casos['mortes'] / (df_soma_casos['mortes'] + df_soma_casos['recuperados']) * 100
df_soma_casos.head()
```

```python
def casos_ate(df, date):
    return df.groupby('Country/Region').sum().iloc[:,2:].loc[:,:date].max(axis = 1)
    
soma_casos_ate_20_2_20 = [casos_ate(casos_confirmados, '2/20/20'), 
                          casos_ate(mortes, '2/20/20'), 
                          casos_ate(recuperados, '2/20/20')]
df_soma_casos_ate_20_2_20 = pd.concat(soma_casos_ate_20_2_20, axis = 1)
df_soma_casos_ate_20_2_20.columns = ['confirmados', 'mortes', 'recuperados']
df_soma_casos_ate_20_2_20
```

```python
# criar os dfs

confirmed = pd.read_csv("https://raw.githubusercontent.com/CSSEGISandData/COVID-19/master/csse_covid_19_data/csse_covid_19_time_series/time_series_covid19_confirmed_global.csv")
deaths = pd.read_csv('https://raw.githubusercontent.com/CSSEGISandData/COVID-19/master/csse_covid_19_data/csse_covid_19_time_series/time_series_covid19_deaths_global.csv')
recovered = pd.read_csv('https://raw.githubusercontent.com/CSSEGISandData/COVID-19/master/csse_covid_19_data/csse_covid_19_time_series/time_series_covid19_recovered_global.csv')
```

```python
confirmed['is_china'] = confirmed['Country/Region'] == 'China'
#.set_index('Country/Region')
```

```python
# confirmed['is_china'].replace(True, 'China', inplace=True)
# confirmed['is_china'].replace(False, 'Outros países', inplace=True)
```

```python
# Usando gráfico de pizza
plt.figure(figsize = (8,7))
confirmed.groupby('is_china').sum()['3/2/20'].plot(kind = 'pie')

plt.title('Gráfico de pizza torna difícil a comparação entre as fatias.\nQual % dos casos fora da China?', fontsize = 15)

plt.show()
```

```python
# Usando gráfico de barra
plt.figure(figsize = (8,7))
confirmed.groupby('is_china').sum()['3/2/20'].plot(kind = 'bar')

plt.title('''
Gráfico de barra proporciona a facilidade na comparação.
China tem, aproximadamente, 8x os casos do resto do mundo.
02/03/2020
''', fontsize = 15)
plt.grid(True, alpha = .4)

plt.show()
```

```python
import numpy as np
```

```python
confirmed.head()
```

```python
confirmed['is_china_label'] = np.where(confirmed['is_china'], 'China', 'Others')
display(confirmed.head())
display(confirmed.tail())
```

```python
# Usando gráfico de barra
plt.figure(figsize = (8,7))
plt.bar(['China', 'Outros países'],
        [confirmed.groupby('is_china_label').sum()['3/2/20']['China'], 
         confirmed.groupby('is_china_label').sum()['3/2/20']['Others']])


plt.title('Gráfico de barra proporciona a facilidade na comparação.\nA China tem, aproximadamente, 8x os casos do resto do mundo.', fontsize = 15)
plt.grid(True, alpha = .4)

plt.show()
```

```python
summed = confirmed.groupby('is_china_label').sum()
new = summed['3/2/20'] - summed['3/1/20']

plt.figure(figsize = (8,7))

new.plot(kind= 'bar')

plt.title('Novos casos em 02/03/2020', fontsize = 15)
plt.grid(True, alpha = .4)

plt.show()
```

```python
summet_T = summed.loc[:,'1/22/20':'3/2/20'].T

# plt.figure(figsize = (8,7))

summet_T.plot()


plt.title('Casos confirmados acumulados até 02/03/2020', fontsize = 15)
# plt.grid(True, alpha = .4)

plt.show()
```

```python
differences = summed.loc[:,'1/22/20':'3/2/20'].T.diff().dropna()
differences
```

```python
plt.figure(figsize = (12,6))
differences.plot()

plt.title('Novos casos até 02/03/2020', fontsize = 15)
# plt.grid(True, alpha = .4)

plt.show()
```
