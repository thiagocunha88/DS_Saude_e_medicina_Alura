```python
un_population = pd.read_csv('https://population.un.org/wpp/Download/Files/1_Indicators%20(Standard)/CSV_FILES/WPP2019_TotalPopulationBySex.csv')
un_population
```

```python
# População vem em milhares, necessário multiplicar por 1000 para ter o valor correto

un_population.loc[:,'PopMale':] = un_population.loc[:,'PopMale':] * 1000
un_population
```

```python
un_population.query('Location.str.contains("United") and Time == 2019', engine='python')
```

```python
un_population_total = un_population.query('Time == 2019')[['Location', 'PopTotal']].set_index('Location')
un_population_total
```

```python
# Padronizar os nomes para que o join funcione e seja possível levar a população ao dataset
# Navio de cruzeiro Diamond Princess tinha 3700 passageiros

def rename_location(location):
    if location == 'Republic of Korea':
        return 'Korea, South'
    if location == 'Iran (Islamic Republic of)':
        return 'Iran'
    if location == 'United States of America':
        return 'US'
    return location

un_population_total['location_by_who'] = un_population_total.index.map(rename_location)
un_population_total.query('location_by_who == "US"')
```

```python
# Colocar a coluna location_by_who como índice para executar o join

un_population_total = un_population_total.set_index('location_by_who')
un_population_total.head()
```

```python
# Refazer o dataset com os dados calculo de letalidade, filtrando pelas datas utilizadas no curso.
# Ideia é inserir neste dataset a população total de cada país.

date_start = '1/22/20'
date_end = '3/3/20'

count_cases = [confirmed.groupby('Country/Region').sum().loc[:,date_start:date_end].max(axis = 1),
               deaths.groupby('Country/Region').sum().loc[:,date_start:date_end].max(axis = 1),
               recovered.groupby('Country/Region').sum().loc[:,date_start:date_end].max(axis = 1)]
df_count_cases = pd.concat(count_cases, axis = 1)
df_count_cases.columns = ['confirmed', 'deaths', 'recovered']

df_count_cases['letality_1'] = df_count_cases['deaths'] / df_count_cases['confirmed'] * 100
df_count_cases['letality_2'] = df_count_cases['deaths'] / (df_count_cases['deaths'] + df_count_cases['recovered']) * 100

df_count_cases
```

```python
df_count_cases_expanded = df_count_cases.join(un_population_total).sort_values('confirmed', ascending=False).dropna()
df_count_cases_expanded.head(20)
```

```python
# Calcular a taxa de incidência e a taxa de mortalidade
# Cálculo errado, pois foi considerada que a população em risco é de todo o país da China,
# porém, grande parte dos casos (83% aproximadamente) são em Hubei.
# Torna o resultado abaixo injusto, pois a população de Hubei é bem menor que da China toda.


df_count_cases_expanded['incidence_rate'] = df_count_cases_expanded.confirmed / df_count_cases_expanded.PopTotal * 100000
df_count_cases_expanded['mortality_rate'] = df_count_cases_expanded.deaths / df_count_cases_expanded.PopTotal * 100000
df_count_cases_expanded.head(10)
```

```python
# FOnte https://pt.wikipedia.org/wiki/Hubei

hubei_population = 59020000
confirmed_cases =  confirmed.set_index('Province/State').loc['Hubei'][date_end]
death_cases =  deaths.set_index('Province/State').loc['Hubei'][date_end]
recovered_cases =  recovered.set_index('Province/State').loc['Hubei'][date_end]

letality_rate_1 = death_cases / confirmed_cases * 100
letality_rate_2 = death_cases / (death_cases + recovered_cases) * 100

print(f'Letality rate 1: {letality_rate_1:.2f}%.')
print(f'Letality rate 2: {letality_rate_2:.2f}%.')
```
